```python
import pandas as pd
import numpy as np
```

```python
## read all the csv files
books = pd.read_csv('Books.csv')
ratings = pd.read_csv('Ratings.csv')
users = pd.read_csv('Users.csv')
```

```python
books.head()
```

## Popularity Based Recommendation System

### Merge books and rating based on ISBN

```python
ratings_with_name = ratings.merge(books,on='ISBN')
ratings_with_name.head()
```

### 1. check no of votes on every book by name

```python
num_ratings = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_ratings.rename(columns={
    'Book-Rating':'num_ratings'
},inplace=True)
num_ratings.head()
```

### 2.Creating a dataframe with book name and respective average ratings

```python
avg_ratings = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_ratings.rename(columns={
    'Book-Rating':'avg_ratings'
},inplace=True)
avg_ratings.head()
```

### 3. Creatig a popular_df dataframe: merging num_raitngs and avg_ratings

```python
popular_df = num_ratings.merge(avg_ratings,on='Book-Title')
popular_df
```

### 4. Only Keeping those records of the books whose voting counts i.e. num_ratings is greater or equal to 250

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False).head(30)
```

```python
popular_df
```

```python
popular_df.shape
```

### Also get the author name , publisher, and image of the book

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
popular_df
```

## Collaborative Filtering Based Recommendation System

```python
ratings_with_name
```

### 1. User with book rating count --> how many times a particular user has voted/rated a book

```python
x = ratings_with_name.groupby('User-ID')['Book-Rating'].count() > 250
padhe_likhe_users = x[x].index
```

```python
padhe_likhe_users
```

```python
filtered_ratings = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
filtered_ratings
```

### 2. Books with rating grater or equal to 50

```python
y = ratings_with_name.groupby('Book-Title')['Book-Rating'].count() > 50
```

```python
famous_books = y[y].index
```

```python
famous_books
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
final_ratings[final_ratings.duplicated()]  ## no duplicated records
```

## Pitvot Table of Users and Books

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

### 3. Calculate the cosine similarities 

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    ## fetch_index
    index = np.where(pt.index == book_name)[0][0]

    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:11]

    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)
    print(data)
```

```python
recommend('Harry Potter and the Prisoner of Azkaban (Book 3)')
```
